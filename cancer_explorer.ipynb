```python
import pandas as pd
```

```python
df = pd.read_csv('californiahealthmaps_county_all.csv')
```

```python
df.head()
```

```python
missing_values = df.isna().sum()
missing_values
```

```python
# Remove rows with missing AAIR
df = df.dropna(subset=['AAIR'])
```

```python
df['Cancer'].unique()
```

```python
df['Cancer'].value_counts()
```

```python
df['Cities'].value_counts()
```

```python
df['Counties'].unique()
```

```python
df = df[(df['Sex'] != "Both") & (df['Years'] == "05yr")]
df
```

```python
df = df[['Counties', 'Sex', 'Cancer', 'PopTot', 'AAIR']]
# Calculate the number of cancer incidents
df.loc[:, 'Cancer_Incidents'] = (df['AAIR'] / 100000) * df['PopTot']
df
```

```python
# Save the DataFrame to a CSV file
df.to_csv('cleaned_cancer_dataset.csv', index=False)
```
