```python
import pandas as pd
```

```python
df = pd.read_csv('../data/raw/californiahealthmaps_county_all.csv')
```

```python
df.head()
```

```python
missing_values = df.isna().sum()
missing_values
```

```python
# Remove rows with missing AAIR
df = df.dropna(subset=['AAIR'])
```

```python
df = df[(df['Sex'] != "Both") & (df['Years'] == "05yr")]
df
```

```python
df = df[['Counties', 'Sex', 'Cancer', 'PopTot', 'AAIR']]
# Calculate the number of cancer incidents
df.loc[:, 'Cancer_Incidents'] = (df['AAIR'] / 100000) * df['PopTot']
df
```

```python
# California Populations from https://www.california-demographics.com/counties_by_population

population_dict = {
    "Tulare": 479468,
    "Fresno": 1017162,
    "Shasta": 180366,
    "Tehama": 64896,
    "Ventura": 829590,
    "Kern": 913820,
}

combined_counties = {
    'Tulare, Fresno': ['Tulare', 'Fresno'],
    'Shasta, Tehama': ['Shasta', 'Tehama'],
    'Ventura, Kern': ['Ventura', 'Kern']
}

new_rows = []

for combined_county, counties in combined_counties.items():
    combined_rows = df[df['Counties'] == combined_county]

    total_population = sum(population_dict[county] for county in counties)
    total_incidents = combined_rows['Cancer_Incidents'].sum()

    for _, row in combined_rows.iterrows():
        for county in counties:
            new_row = row.copy()
            new_row['Counties'] = county
            new_row['PopTot'] = population_dict[county]
            new_row['Cancer_Incidents'] = (new_row['PopTot'] / total_population) * row['Cancer_Incidents']
            new_row['AAIR'] = (new_row['Cancer_Incidents'] / new_row['PopTot']) * 100000
            new_rows.append(new_row)

new_df = pd.DataFrame(new_rows)
df = df[~df['Counties'].isin(combined_counties.keys())]
final_df = pd.concat([df, new_df], ignore_index=True)
# Rename Counties column to County
final_df = final_df.rename(columns={'Counties': 'county'})

final_df











```

```python
# Save the DataFrame to a CSV file
final_df.to_csv('../data/cleaned/cleaned_cancer_dataset.csv', index=False)
```
